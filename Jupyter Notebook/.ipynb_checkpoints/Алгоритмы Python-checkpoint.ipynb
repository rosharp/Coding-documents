def pow(a=float, n:int):
    if n == 0:
        return 1
    elif n % 2 == 1: # Нечет
        return pow(a, n - 1) * a
    else: # n - четное
        return pow(a**2, n // 2)

#### Ханойские башни 

**Нужно переложить с первого столбца на второй всю пирамидку. Нельзя хватать сразу несколько, лежать они должны по размеру по порядку. 

## def merge(A:list, B:list):
    C = [0] * (len(A) + len(B)) # Сливание двух длин списков
    """ i - индекс для списка А, к - индекс для списка В, n - индекс для 
        списка С. Массивы отсортированы по возрастанию. Если А[i] меньше 
        B[k], то i увеличивается на 1 и перешагивает на следующий элемент, 
        n также увеличивается на 1. Элементы сравниваются до тех пор, пока
        один из массивов не закончится. 
        
        Сортировка назвается устойчивой, если она не меняет порядок равных 
        элементов. Не все элементы, которые равны друг другу, абсолютно
        одинаковы. Сортируемая характеристика может быть не единственной. 
    """
    i = k = n = 0
    while i < len(A) and k < len(B): # Элементы не вышли за границы
        if A[i] <= B[k]:
            C[n] = A[i]; i += 1; n += 1
        else:
            C[n] = B[k]; k += 1; n += 1 
    while i < len(A):
        C[n] = A[i]; i += 1; n += 1
    while k < len(B):
        C[n] + B[k]; k += 1; n += 1
    return C

```python
# Рекурсивная функция. 

""" B = [5, 2, 7, 3, 1]
    merge.sort(B)
    print(*B)
    
    B --> [5, 2, 7, 3, 1]
    Работа с этим списком под именем А. 
    В процессе работы два новых объекта L, R, C. 
    Если А = С, то ссылки поменялись местами. 
    Но при этом исходное имя В не изменилось. 
    После этого A, L, R, C будут уничтожены. 
"""
def merge_sort(A): # Принимает массив А для сортировки 
    if len(A) <= 1:
        return
    middle = len(A) // 2
    L = [A[i] for i in range(0, middle)]
    R = [A[i] for i range(middle, len(A))]
    merge_sort(L)
    merge_sort(R)
    C = merge(L, R)
    for i in range(len(A)):
        A[i] = C[i]
```

```python
# Сортировка Тони Хоара (QuickSort)

""" Массив А разбиваю, но не напополам, и выбираю барьерный элемент. 

    A
    
    barirer = A[0]
    
    L -- M -- R 
    
    Переливание из барьева в один из L, M, R, а потом их переливание обратно в А. 
"""

def hoar_sort(A): 
    if len(A) <= 1:
        return None # Крайний случай. Имена существуют, пока существует их контекст.
    barrier = A[0]; L = []; M = []; R = []
    for x in A:
        if x < barrier:
            L.append(x)
        elif x == barrier:
            M.append(x)
        else:
            R.append(x)
    hoar_sort(L) # В массиве М элементы равны, сортировать не надо
    hoar_sort(R)
    k = 0
    for x in L + M + R: # Объединенный конкатенированием массив 
        A[k] = x; k += 1
```

```python
def check_sortred(A, ascending=True): # Функция сортирована по возрастанию
    """ Проверка отсортированности за O(len(A))
    
    int(True) = 1 -> +1
    int(False) = 0 -> -1
    S = 2*x-1, x=int(ascending)
    """
    flag = True; S = 2 * int(ascending - 1)
    for i in range(0, N-1):
        if S * A[i] > S * A[i+1]:
            flag = False
            break
    return flag
```

```python
# Бинарный поиск в массиве. 

""" Сортировать надо, чтобы быстро искать. 
    [1, 2, 2, 2, 3, 4, 5, *5*, 5, 5, 5, 7, 7, 7, 7]
    left_bound 
    right_bound
    Функция, которая будет находить левые и правые границы, а также выход для 
    несуществующих чисел. left_bound на элемент, который меньше, чем 6. 
    right_bound будет указывать на больший, чем 6. Для 7 right_bound
    будет len всего массива. Не обязан указывать на элемент массива вообще. 
    8 или 10 - right_bound снова len, а left_bound будет указывать послед-
    нее правое число. 
    
    Требование: массив отсортирован. 
    Из функции мы хотим получить индекс. Левая и правая границы, 
    которые укажут на элемент. Левая - на элемент, который меньше 
    текущего, а правая - на тот, который больше числа. 
    
    1-*3*-*3*-6-7-9
    x = 5
    left_boundary
    right_boundary

"""

def left_bound(A, key): # Массив - А
    left = -1
    right = len(A)
    while right - left > 1:
        middle = (left + right) // 2
        if A[middle] < key:
            left = middle 
        else:
            right = middle
    return left
```

```python
def right_bound(A, key): # Массив - А
    left = -1
    right = len(A)
    while right - left > 1:
        middle = (left + right) // 2
        if A[middle] <= key:
            left = middle 
        else:
            right = middle
    return right

# O(log2N)
```

```python
# Наибольшая возрастающая последовательность

""" F:  = НВП для части А(среза)[0:i], которая заканчивается
    и содержит элемент ai=A[i-1]
    
    [a1, a2, a3... ai-1, ai] длина наибольшей возрастающей
    подпоследовательности
         [a1, a2...aj] можно ли продолжить с ai? только если
         ai больше, чем aj. aj < ai. 
         
    Fi = max(Fj) + 1
         j < i
         ai > aj # иначе последовательность не является возр.
         
    F0 = 0 # крайний случай 
"""

def gis(A):
    F = [0] * (len(A) + 1)
    for i in range(1, len(A) + 1):
        m = 0
        for j in range(0, j):
            if A[i] > A[j] and F[j] > m:
                m = F[j]
    F[i] = m + 1
return F[len(A)]
```

Что такое ассимптотика? 
Выборка чисел: прапорщик может пройтись 1 раз (10 операций по 10 числам), а может 20 - сортировка пузырьком. Среднестатистическая, худший случай. 

Наихудший и среднестатистический могут вести себя по-разному. 

Начиная с n[0] будет покрывать сверху n[max][n], n - масштааб задач. nmax[0n] = 0(f(n)).

0(1) - фиксированное время, не зависящее от количества чисел
0(log2N) - основание часто не пишут. Например, метод поиска среди упорядоченных чисел делением пополам. 
O(N) - однопроходный алгоритм, ка ждое число надо посмотреть. 
O(N+M) - M - диапазон, количество возможных чисел. Бывает несколько параметров масштааб. 
O(N*log N) - быстрая сортировка Хоара, сортировка слияния
O(N^2) - пузырек, вставками, выбором 
O(N^3) - специфическая сортировка дурака. 
O(N!)
O(2^N) - комбанаторые ситуации полного перебора.

#### Задача о дискретном рюкзаке

Метод динамического программирования. Массы и стоимости будут дискретными числами. 

F(k, i) - максимальная стоимость предметов, которые можно уложить в рюкзак емкостью K. 

При этом можно использовать только первые i предметов. Двумерное динамическое программирование, два параметра: взять-не взять.

M - максимально допустимая емкость рюкзака. 

```python
# m[i] - масса, V[i] - стоимость
F = [[0] * (N + 1) for i in range(M+1)] # N - предметы
for i in range(1, n+1):
    k in range(1, M): 
        if m[i] <= k:
            F[k][i] = max(F[i-1][k], V[i] + F[k-m[i]][i-1])
        else:
            F[k][i] = F[k][i-1]
# F[M][N] - ответ
```
